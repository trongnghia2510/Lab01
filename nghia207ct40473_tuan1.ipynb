```python
#Import thư viện pandas: Script import thư viện pandas với tên viết tắt là 'pd'. 
import pandas as pd

# Đọc dữ liệu từ một tập tin CSV vào một DataFrame của pandas.
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path)

#Tính điểm trung bình môn (TBM) cho từng năm học (TBM1 cho năm 1, TBM2 cho năm 2 và TBM3 cho năm 3) bằng cách lấy trung bình 
# của các cột tương ứng.
cols = ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']
for i in range(1, 4):
    df[f'TBM{i}'] = df[[f'{col}{i}' for col in cols]].mean(axis=1)
    
#Định nghĩa một hàm có tên xep_loai để xác định xếp loại dựa trên điểm trung bình môn (TBM) đầu vào.
def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif tbm < 6.5:
        return 'TB'
    elif tbm < 8.0:
        return 'K'
    elif tbm < 9.0:
        return 'G'
    return 'XS'
#Tạo các cột mới 'XL1', 'XL2', 'XL3' trong DataFrame và gán kết quả xếp loại từ việc áp dụng hàm 
# xep_loai cho các cột TBM1, TBM2, TBM3 tương ứng.
for i in range(1, 4):
    df[f'XL{i}'] = df[f'TBM{i}'].apply(xep_loai)
    
#Lưu DataFrame đã cập nhật với các biến xếp loại vào cùng tập tin CSV ban đầu.
df.to_csv(file_path, index=False)

#In ra thông báo "Đã tạo và lưu các biến xếp loại thành công!" để thông báo hoàn thành.
print("Đã tạo và lưu các biến xếp loại thành công!")
```

```python
#Import thư viện pandas với tên viết tắt là pd.
import pandas as pd

# Đọc dữ liệu từ một tập tin CSV vào một DataFrame của pandas.
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path)

#Định nghĩa danh sách các cột chứa điểm thi theo từng môn (T, L, H, S, V, X, D, N).
cols = ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']
#Sử dụng một vòng lặp để tính điểm trung bình môn (TBM) cho mỗi năm học (1 đến 3) 
# bằng cách lấy trung bình của các cột tương ứng (TBM1, TBM2, TBM3).
for i in range(1, 4):
    df[f'TBM{i}'] = df[[f'{col}{i}' for col in cols]].mean(axis=1)
    
#Định nghĩa một hàm có tên là Chuyen_Sang_Diem_USA để chuyển đổi điểm từ thang điểm 10 về thang điểm 4 của Mỹ.
def Chuyen_Sang_Diem_USA(tbm):
    return (tbm / 10) * 4

#Sử dụng một vòng lặp khác để áp dụng hàm Chuyen_Sang_Diem_USA 
# cho các cột TBM1, TBM2, TBM3 và lưu kết quả vào các cột mới có tiền tố 'US_'.
for i in range(1, 4):
    df[f'US_TBM{i}'] = df[f'TBM{i}'].apply(Chuyen_Sang_Diem_USA)
    
#Lưu DataFrame đã cập nhật với các điểm trung bình chuyển đổi vào cùng một file CSV.
df.to_csv(file_path, index=False)

#In ra thông báo "Đã chuyển đổi và lưu các điểm trung bình thành công!" để thông báo hoàn thành.
print("Đã chuyển đổi và lưu các điểm trung bình thành công!")

#Hiển thị 5 dòng đầu tiên của DataFrame sau khi cập nhật.
df.head()
```

```python
#Import thư viện pandas với tên viết tắt là pd.
import pandas as pd

#Đọc dữ liệu từ tập tin CSV vào một DataFrame của pandas.
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path)
#Định nghĩa một hàm có tên là calculate_kqxt để tính toán kết quả xét tuyển (KQXT) cho từng 
# thí sinh dựa trên thông tin về nhóm môn học (KT) và điểm thi (DH1, DH2, DH3) của họ.
def calculate_kqxt(row):
    group = row['KT']
    DH1, DH2, DH3 = row['T1'], row['L1'], row['H1']
#Trong hàm calculate_kqxt, điểm trung bình được tính dựa trên các điểm thi và nhóm môn học. 
# Nếu nhóm môn học là 'A' hoặc 'A1', thì điểm trung bình được tính theo tỷ lệ 2:1:1 cho các môn Toán, Lý, Hóa.
# Nếu nhóm môn học là 'B', thì tỷ lệ sẽ là 1:2:1 cho các môn tương ứng. Trong trường hợp khác, 
# điểm trung bình được tính theo trung bình cộng của ba môn.    
    if group in ['A', 'A1']:
        average_score = (DH1 * 2 + DH2 + DH3) / 4
    elif group == 'B':
        average_score = (DH1 + DH2 * 2 + DH3) / 4
    else:
        average_score = (DH1 + DH2 + DH3) / 3
    return 1 if average_score >= 5.0 else 0 #Hàm trả về 1 nếu điểm trung bình lớn hơn hoặc bằng 5.0 và 0 nếu điểm trung bình nhỏ hơn 5.0.

#Sử dụng phương thức apply() của DataFrame để áp dụng hàm calculate_kqxt cho mỗi hàng (axis=1).
#Tạo một cột mới trong DataFrame có tên là 'KQXT' để lưu kết quả xét tuyển được tính toán.
df['KQXT'] = df.apply(calculate_kqxt, axis=1)

#Lưu DataFrame đã cập nhật với cột 'KQXT' mới vào cùng một file CSV.
df.to_csv(file_path, index=False)

#Hiển thị 5 dòng đầu tiên của DataFrame sau khi cập nhật.
print(df.head())
```
